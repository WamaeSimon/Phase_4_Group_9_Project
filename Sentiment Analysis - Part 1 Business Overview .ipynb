###### Removing urls, https, www

```python
import re
def remove_urls(text, replacement_text= ' url'):
    url_pattern = re.compile(r'https?:\/\/\S+|www\.\S+')
    urls = url_pattern.findall(text)
    
    for url in urls:
        text = text.replace(url, replacement_text)
    return text
# Create new column with tokenized data
df["clean_tweet_text"] = df["tweet_text"].apply(remove_urls)
```

* To implement the RegexpTokenizer for tokenizing individual words in a corpus, you will first need to import it from the nltk tokenize module. The RegexpTokenizer allows you to define a custom regular expression pattern to match word tokens, typically using \w+ to capture alphanumeric word sequences. Once applied to the corpus, this tokenizer splits the text into a list of words, removing punctuation and other non-word characters according to the specified regular expression pattern.

```python
import pandas as pd
from nltk.tokenize import RegexpTokenizer

def tokenize_and_display(df):
    # Define the basic token pattern
    basic_token_pattern = r"(?u)\b\w\w+\b"
    
    # Initialize the tokenizer with the specified pattern
    tokenizer = RegexpTokenizer(basic_token_pattern)
    
    # Apply the tokenizer to a new column in the DataFrame
    df["text_tokenized"] = df["clean_tweet_text"].apply(tokenizer.tokenize)
    
    # Display the DataFrame with left-aligned text
    return df["text_tokenized"]

# Call the function
tokenize_and_display(df)
```

* Notice that the regexpTokenizer removes some special characters, including urls. It splits the specific words and puts them into a list. However it does not remove the stopwords/ filler words and numbers.

```python
## Get the list of English stop words
stop_words = stopwords.words('english')
stop_words += list()
# Function to remove stop words
def remove_stop_words(tokens_list):
    """
    Removes stop words from a list of tokens.
    Arguments:
    - tokens_list: List of tokenized words.
    
    Returns:
    - List of tokens without stop words.
    """
    return [word for word in tokens_list if word.lower() not in stop_words]

# Apply the function to the tokenized text column
df["text_tokenized"] = df["text_tokenized"].apply(remove_stop_words)
df.head()
```

```python
import re

def remove_numbers_and_special_patterns(text):
    # Combine all patterns into a single regex to remove numbers and special patterns
    pattern = r'^\d+[a-zA-Z]+' \
              r'|^[a-zA-Z]+\d+\d+$' \
              r'|^\d+(\.\d+)?$' \
              r'|^[a-zA-Z]+\d+[a-zA-Z]+$'\
              r'[^a-z\s#@]'
    
    
    # Remove the matched patterns and clean up spaces between letters
    cleaned_words = [re.sub(pattern, '',word ) for word in text]
    
    # Join the cleaned words back into a single string without spaces
    result = ' '.join(cleaned_words)
    
    return result

# Assuming 'df' is your DataFrame containing the 'text_tokenized' column with lists of words
df['text_tokenized'] = df['text_tokenized'].apply(remove_numbers_and_special_patterns)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#defining a function for wordcloud generation
def generate_wordcloud(tokens, collocations=False, background_color='black', 
                       colormap='Greens', display=True):
    
    """Function generates and returns a wordcloud based on a tokens list passed in.
    -------------------------------
    Arguments:
    tokens: a tokens list
    collocations: Whether to include collocations (bigrams) of two words
    background_color: background color of the resulting word cloud
    colormap: the color map for the words that will be in the word cloud
    display: Whether to show the resulting wordcloud"""
    
    ## Initalize a WordCloud
    wordcloud = WordCloud(collocations=collocations, 
                          background_color=background_color, 
                          colormap=colormap, 
                          width=500, height=300)

    ## Generate wordcloud from tokens
    wordcloud.generate(' '.join(tokens))

    ## Plot with matplotlib
    if display:
        plt.figure(figsize = (12, 15), facecolor = None) 
        plt.imshow(wordcloud) 
        plt.axis('off');
    return wordcloud
```

##### Binary Classification Model : Using Multinomial Bayes
* The Multinomial Naive Bayes is a machine learning approach that classifies data into two distinct classes based on features.
* The model calculates the likelihood of each class given the observed features and assigns the class with the highest probability. Despite its simplicity, Multinomial Naive Bayes is effective for binary classification problems, particularly when the data is sparse and high-dimensional.

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegressionCV
```

```python
from sklearn.metrics import classification_report, roc_curve
from sklearn.metrics import RocCurveDisplay, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def classification(y_true, y_pred, X_test, X_train, clf):
    """
    This function shows the classification report,
    the confusion matrix as well as the ROC curve for evaluation of model quality.
    -----------------------------------
    Arguments:
    y_true: Correct y values, typically y_test that comes from the 
    train_test_split performed at the beginning of model development.
    y_pred: Predicted y values by the model.
    clf: Classifier model that was fit to training data.
    X_test: X_test values from train_test_split.
    X_train: X_train values from train_test_split.
    """
    
    # Training and test scores
    print(f"Training Score: {round(clf.score(X_train, y_train), 2)} \
            Test Score: {round(clf.score(X_test, y_true), 2)}")
    
    # Classification report
    print("\n")
    print("CLASSIFICATION REPORT")
    print("------------------------------------------")
    print(classification_report(y_true=y_true, y_pred=y_pred))
    
    # Creating a figure/axes for confusion matrix and ROC curve
    fig, ax = plt.subplots(ncols=2, figsize=(12, 5))
    
    # Plotting the normalized confusion matrix
    ConfusionMatrixDisplay.from_estimator(estimator=clf, X=X_test, y=y_true, 
                                          cmap='Blues', normalize='true', 
                                          ax=ax[0], display_labels=['Negative', 'Positive'])
    ax[0].set_title("Normalized Confusion Matrix")

    
    # Plotting the ROC curve
    RocCurveDisplay.from_estimator(estimator=clf, X=X_test, y=y_true, ax=ax[1])
    
    # Plotting the 50-50 guessing plot for reference
    ax[1].plot([0, 1], [0, 1], ls='--', color='orange')
    ax[1].set_title("ROC Curve")
    
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

def classification_with_matrix(y_true, y_pred, X_test, X_train, clf):
    """
    This function shows the classification report, the confusion matrix (both as numbers and a plot),
    and the ROC curve for evaluation of model quality.
    -----------------------------------
    Arguments:
    y_true: Correct y values, typically y_test that comes from the 
    train_test_split performed at the beginning of model development.
    y_pred: Predicted y values by the model.
    clf: Classifier model that was fit to training data.
    X_test: X_test values from train_test_split.
    X_train: X_train values from train_test_split.
    """
    
    # Training and test scores
    print(f"Training Score: {round(clf.score(X_train, y_train), 2)} \
            Test Score: {round(clf.score(X_test, y_true), 2)}")
    
    # Classification report
    print("\n")
    print("CLASSIFICATION REPORT")
    print("------------------------------------------")
    print(classification_report(y_true=y_true, y_pred=y_pred))
    
    # Displaying Confusion Matrix Numbers
    cm = confusion_matrix(y_true, y_pred)
    print("\nCONFUSION MATRIX")
    print("------------------------------------------")
    print(f"True Negatives: {cm[0, 0]}")
    print(f"False Positives: {cm[0, 1]}")
    print(f"False Negatives: {cm[1, 0]}")
    print(f"True Positives: {cm[1, 1]}")
    print("\nConfusion Matrix Table:\n", cm)
```
