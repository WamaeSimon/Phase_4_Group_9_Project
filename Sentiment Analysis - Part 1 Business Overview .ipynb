**Sentiment Analysis** also known as **Sentiment Classification** in brief uses natural language processing to identify the emotional tone behind text, such as customer feedback, and categorize it as positive, negative, or neutral. 

The above can also be described as a text classification tasks, where we look at a phrase, or a list of phrases and use a classifier to tell if the sentiment behind that is:
- positive
- negative 
- neutral.

###  1.4 **Metrics of Success**

To evaluate the success of our sentiment analysis model, we will use metrics such as; accuracy, precision, recall or sensitivity, f1 score and the confusion matrix. To evaluate the performance of the sentiment classification model, we will use the following metrics:
#### 1. **Accuracy**:
   - Accuracy will check at the percentage of the correctly classified instances of sentiments out of the total sentmental            instances.
   - Target: **85% or higher**.

#### 2. **Precision**:
   - Precision will tell the percentage of actually correct positive sentiment predictions, thus telling us how often the model      is correct when it predicts a positive sentiment. The percentage of actual positive sentiments, that are correctly              identified by the model will be shown by recall. This metrics is important to strike a tradeoff between true positives and      false negatives
   - Target: **80% or higher** for each class (positive, negative, neutral).

#### 3. **Recall**:
   - Measure the model’s ability to correctly identify all relevant examples of a specific sentiment.
   - Target: **75% or higher** for each class.

#### 4. **F1-Score**:
   - Provide a balanced metric that considers both precision and recall.
   - Target: **80% or higher** overall.

#### 5. **Business Impact**:
   - Improved customer satisfaction through the identification of key negative sentiments.
   - Better marketing strategies based on trends in positive feedback.

```python
import re
def remove_urls(text, replacement_text= ' url'):
    url_pattern = re.compile(r'https?:\/\/\S+|www\.\S+')
    urls = url_pattern.findall(text)
    
    for url in urls:
        text = text.replace(url, replacement_text)
    return text
# Create new column with tokenized data
df["clean_tweet_text"] = df["tweet_text"].apply(remove_urls)
```

```python
import re

def remove_numbers_and_special_patterns(text):
    # Combine all patterns into a single regex to remove numbers and special patterns
    pattern = r'^\d+[a-zA-Z]+' \
              r'|^[a-zA-Z]+\d+\d+$' \
              r'|^\d+(\.\d+)?$' \
              r'|^[a-zA-Z]+\d+[a-zA-Z]+$'
    
    # Remove the matched patterns and clean up spaces between letters
    cleaned_words = [re.sub(pattern, '',word ) for word in text]
    
    # Join the cleaned words back into a single string without spaces
    result = ' '.join(cleaned_words)
    
    return result

# Assuming 'df' is your DataFrame containing the 'text_tokenized' column with lists of words
df['text_tokenized'] = df['text_tokenized'].apply(remove_numbers_and_special_patterns)
```

```python
## Get the list of English stop words
stop_words = stopwords.words('english')
stop_words += list()
# Function to remove stop words
def remove_stop_words(tokens_list):
    """
    Removes stop words from a list of tokens.
    Arguments:
    - tokens_list: List of tokenized words.
    
    Returns:
    - List of tokens without stop words.
    """
    return [word for word in tokens_list if word.lower() not in stop_words]

# Apply the function to the tokenized text column
df["text_tokenized"] = df["text_tokenized"].apply(remove_stop_words)
df.head()
```

Create a function that we call nd returns the word clouds for specified data set.
The function takes in the tokenized column and displays the word cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#defining a function for wordcloud generation
def generate_wordcloud(tokens, collocations=False, background_color='black', 
                       colormap='Greens', display=True):
    
    """Function generates and returns a wordcloud based on a tokens list passed in.
    -------------------------------
    Arguments:
    tokens: a tokens list
    collocations: Whether to include collocations (bigrams) of two words
    background_color: background color of the resulting word cloud
    colormap: the color map for the words that will be in the word cloud
    display: Whether to show the resulting wordcloud"""
    
    ## Initalize a WordCloud
    wordcloud = WordCloud(collocations=collocations, 
                          background_color=background_color, 
                          colormap=colormap, 
                          width=500, height=300)

    ## Generate wordcloud from tokens
    wordcloud.generate(' '.join(tokens))

    ## Plot with matplotlib
    if display:
        plt.figure(figsize = (12, 15), facecolor = None) 
        plt.imshow(wordcloud) 
        plt.axis('off');
    return wordcloud
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
```

```python
# Step 1: Convert the text data to TF-IDF features
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(df_model['text_tokenized_lemmatized'])

# X_tfidf is now a sparse matrix with TF-IDF features
print(X_tfidf.shape)  # Check the shape of the TF-IDF matrix
```

### Train the Clustering Model (KMeans)
In this step, we’ll use the KMeans algorithm to cluster the data. Specify the number of clusters (n_clusters) and fit the model.

```python
# Apply KMeans clustering on the TF-IDF data
from sklearn.cluster import KMeans

# Defining the number of clusters
n_clusters = 2 

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
# Training the model
y_kmeans = kmeans.fit_predict(X_tfidf)

## Print the cluster assignments for each document
print(f"Cluster Assignments: {y_kmeans}")
```

### Predict Cluster Assignments
Once the model is trained, we can use it to predict the cluster assignment for each data point. These predictions are already generated by fit_predict().

```python
# Step 3: Predict the cluster assignments for each document
predictions = kmeans.predict(X_tfidf)

# Print the predictions
print(f"Predicted Cluster Assignments: {predictions}")
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Step 4: Evaluate the model using different metrics

# Inertia
print(f"Inertia: {kmeans.inertia_}")

# Silhouette Score (higher is better)
silhouette_avg = silhouette_score(X_tfidf, y_kmeans)
print(f"Silhouette Score: {silhouette_avg}")

# Davies-Bouldin Index (lower is better)
db_index = davies_bouldin_score(X_tfidf.toarray(), y_kmeans)
print(f"Davies-Bouldin Index: {db_index}")

from sklearn.metrics import adjusted_rand_score

ari = adjusted_rand_score(df_model['Emotion_encoded'], y_kmeans)
print(f"Adjusted Rand Index: {ari}")
```

### Evaluating the model

```python
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score

homogeneity = homogeneity_score(df_model['Emotion_encoded'], y_kmeans)
completeness = completeness_score(df_model['Emotion_encoded'], y_kmeans)
v_measure = v_measure_score(df_model['Emotion_encoded'], y_kmeans)

print(f"Homogeneity: {homogeneity}")
print(f"Completeness: {completeness}")
print(f"V-Measure: {v_measure}")
```

### Visualize the Clusters
Since the data is high-dimensional, it’s helpful to use dimensionality reduction to project the data to 2D for visualization. The tfidf vectorizer returns sparse matrix. To reduce the dimensionality, implement the dimensionality algorithm that work well with the sparse matrices. for this case use Truncated SVD. It is a PCA like algorithm for sparse matrices. Parse in the n_components = 100 to retain as much of the explained variation.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Step 3: Apply PCA to reduce the data to 2D for visualization
pca = TruncatedSVD(n_components=90)
X_pca = pca.fit_transform(X_tfidf.toarray())  # Convert to dense array if sparse

# Step 4: Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', s=100, alpha=0.7)
plt.title(f'KMeans Clustering with {n_clusters} Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Optionally: Plot the cluster centers in the PCA space
centers = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centroids')

plt.legend()
plt.show()
```
